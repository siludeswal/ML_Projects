```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
X = df.drop('Wine',axis=1)
y = df["Wine"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=101,test_size=0.3)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x_train.iloc[:,0],x_train.iloc[:,1],c=y_train,cmap="coolwarm",edgecolors="k")
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.show()
```

```python
model_qda = QuadraticDiscriminantAnalysis()
model_qda.fit(x_train,y_train)
model_qda
```

```python
model_qda_predict = model_qda.predict(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_tr_sc = sc.fit_transform(x_train)
x_te_sc = sc.transform(x_test)

from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression(max_iter=1000,C=1.0)
model_log.fit(x_tr_sc,y_train)
model_log
```

```python
model_log_pred = model_log.predict(x_te_sc)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix,accuracy_score,classification_report
print("QDA model Analyis: ")
print(f"Acccuracy score : {accuracy_score(y_test,model_qda_predict)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test,model_qda_predict)}")
print(f"Classification Report: \n{classification_report(y_test,model_qda_predict)}")
ConfusionMatrixDisplay(confusion_matrix(y_test,model_qda_predict)).plot()
print()
```

```python
print("Logistic model Analyis: ")
print(f"Acccuracy score : {accuracy_score(y_test,model_log_pred)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test,model_log_pred)}")
print(f"Classification Report: \n{classification_report(y_test,model_log_pred)}")
ConfusionMatrixDisplay(confusion_matrix(y_test,model_log_pred)).plot()
```

```python
from joblib import dump
dump(model_log,"final_model.joblib")
```
