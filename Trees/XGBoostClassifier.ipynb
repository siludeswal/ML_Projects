# importing necessary libraries

```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,classification_report
```

# Loading dataset

```python
data = load_breast_cancer(as_frame=True)
df = data.frame
df.head()
```

# Exploring dataset

```python
df.info()
```

```python
df.shape
```

# Extracting target and feature variables

```python
X = df.drop('target',axis=1)
y = df['target']
```

# splitting dataset for training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=101,test_size=0.15)
```

# Base XGBClassifier

```python
model = XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

# Finding best hyper parameters values

```python
model = XGBClassifier()
```

```python
param = {
    'n_estimators':[60,80,100,120,150,200],
    'learning_rate':[0.001,0.01,0.1,1,10],
    'max_depth':[1,2,3,4],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1]

}
```

```python
grid = GridSearchCV(estimator=model,param_grid=param,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

# model with best parameters

```python
best_model = XGBClassifier(n_estimators=60,learning_rate = 2,max_depth = 2,colsample_bytree= 0.8,subsample=1)
best_model.fit(x_train,y_train)
```

```python
y_pred = best_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
conf = confusion_matrix(y_test,y_pred)
```
