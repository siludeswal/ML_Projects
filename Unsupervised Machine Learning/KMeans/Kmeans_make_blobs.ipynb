# Importing necessary Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_blobs(n_samples=300, centers=4, n_features=2, random_state=42)

df = pd.DataFrame(X, columns=['C1', 'C2'])
df.head()
```

```python
inertia = []
score = []
for i in range(2,11):
  model = KMeans(max_iter=1000,n_clusters=i)
  labels = model.fit_predict(df)
  inertia.append(model.inertia_)
  score.append(silhouette_score(X,labels))
```

```python
plt.plot(range(2,11),inertia,marker = 'o')
plt.xlabel('k values')
plt.ylabel('inertia')
plt.title('Elbow method')
plt.show()
```

```python
final_model = KMeans(n_clusters=4,max_iter=1000)
labels = final_model.fit_predict(df)
labels
```

```python
sns.scatterplot(data = df,x = 'C1',y = 'C2',hue = labels,palette='Set1',s = 60)
```
