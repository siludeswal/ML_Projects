```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
```

```python
N = 87
x = torch.randn(N,1)
y = x + torch.randn(N,1)/2
```

```python
plt.plot(x,y,'s',color="red")
```

```python
model = nn.Sequential(
    nn.Linear(1,8),
    nn.ReLU(),
    nn.Linear(8,1)
)
```

```python
loss_fun = nn.MSELoss()
learning_rate = 0.05
epochs = 500
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)
loss = torch.zeros(epochs)
```

```python
for epoch in range(epochs):
    y_HAT = model(x)
    losses = loss_fun(y_HAT,y)
    loss[epoch] = losses
    optimizer.zero_grad()
    losses.backward()
    optimizer.step()

prediction = model(x)
```

```python
final_loss = loss_fun(prediction,y)
```

```python
plt.scatter(x,y,color="red",label = "Real Data")
plt.scatter(x,prediction.detach(),color="green",label = "Predictions",marker="*")
plt.legend()
plt.title("Real Data vs Predictions")
plt.show()
```

```python
plt.plot(loss.detach(),"o")
plt.plot(epochs,final_loss.detach(),"o")
plt.show()
```
